```python
import pandas as pd
import os
print(os.getcwd())
filepath = "C:/Users/Giuseppe/3_PYTHON/FILE_ESERC_CSV/boston.csv"
boston = pd.read_csv(filepath)
```

```python
bos_f1= (boston['chas']==1) #La media del prezzo delle case cambia a seconda della distanza dal fiume...]
bos_f2= (boston['chas']==0) # Ho fatto 2 filtri in relazione alla colonna 'chas' che indica la distanza dal fiume#
```

```python
med_1_vic_fiume=boston[bos_f1].groupby('chas') ['medv'].mean().round(2) #ho fatto i raggruppamenti e calcolato la media per i 2 filtri #
med_1_lon_fiume=boston[bos_f2].groupby('chas') ['medv'].mean().round(2) #--- bisogna aggiustare la stampa ---
print(f'La media del costo delle case in relazione alla distanza dal fiume Charles e:\nvicino :{med_1_vic_fiume}\nlontano {med_1_lon_fiume}')
```

```python
#Si nota una correlazione tra tasso di criminalità e valore delle abitazioni? Come si può spiegare?#
boston['crim'].describe()
```

```python
q1=boston['crim'].quantile(0.25)
f_q1=boston['crim']<=q1
boston_crim_1=boston[f_q1]
boston_crim_1
q2=boston['crim'].quantile(0.50)
f_q2=(boston['crim']>q1) & (boston['crim']<=q2)
boston_crim_2=boston[f_q2]
boston_crim_2
q3=boston['crim'].quantile(0.75)
f_q3=(boston['crim']>q2) & (boston['crim']<=q3)
boston_crim_3=boston[f_q3]
boston_crim_3
f_q4=(boston['crim']>q3)
boston_crim_4=boston[f_q4]
boston_crim_4
```

```python
m1=boston_crim_1['medv'].mean().round(2)
print(f' La media del 1° quartile del tasso di criminalità è : {m1}')
m2=boston_crim_2['medv'].mean().round(2)
print(f' La media del 2° quartile del tasso di criminalità è : {m2}')
m3=boston_crim_3['medv'].mean().round(2)
print(f' La media del 3° quartile del tasso di criminalità è : {m3}')
m4=boston_crim_4['medv'].mean().round(2)
print(f' La media del 4° quartile del tasso di criminalità è : {m4}')
```
